## Checkboard


```python
import numpy as np
import matplotlib.pyplot as plt

imageSize = 1024
step = 1024//16

checkboard = np.ones((imageSize,imageSize,3))
for i in range(0, imageSize, step):
    for j in range(0, imageSize, step):
        if (i + j) % (step*2) == 0:
            checkboard[i: i+step, j: j+step] = [1/imageSize * i, 0, 1/imageSize * j]

plt.title("Checkboard")
plt.axis('off')
plt.imshow(checkboard)
plt.show
```

## Checkboard Histogram

```python
mean = np.mean(checkboard)
std_dev = np.std(checkboard)

print(f"Mean: {mean}")
print(f"Standard Deviation: {std_dev}")

plt.figure()
plt.hist(checkboard.ravel(), bins=256, color='grey', alpha=1)
plt.title('Checkboard Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

## Warped Checkboard

```python
x = np.linspace(0, 2 * np.pi, imageSize)
y = np.linspace(0, 2 * np.pi, imageSize)
xx, yy = np.meshgrid(x, y)

warp_x = (xx + 0.5 * np.sin(yy)) * (imageSize/(2*np.pi))
warp_y = (yy + 0.5 * np.sin(xx)) * (imageSize/(2*np.pi))

warp_x = np.clip(warp_x, 0, imageSize-1).astype(int)
warp_y = np.clip(warp_y, 0, imageSize-1).astype(int)

warped_checkboard = checkboard[warp_x, warp_y]

plt.imshow(warped_checkboard)
plt.title("Warped Checkboard")
plt.axis('off')
plt.show()
```
